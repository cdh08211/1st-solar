```python
import keras
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('c:/jupyter/data2.csv')
df.head()
```

```python
x = df.drop(columns=['solar energy production', 'year'])
y = df[['solar energy production']]
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(8, activation='relu', input_shape=(8,)))
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(1))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(x, y, epochs=60, callbacks=[keras.callbacks.EarlyStopping(patience=5)])
```
